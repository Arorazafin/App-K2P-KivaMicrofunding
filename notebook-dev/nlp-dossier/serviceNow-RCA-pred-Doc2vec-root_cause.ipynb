## Text preprocessing

```python
import re
from bs4 import BeautifulSoup
def cleanText(text):
    text = BeautifulSoup(text, "lxml").text
    text = re.sub(r'\|\|\|', r' ', text) 
    text = re.sub(r'http\S+', r'<URL>', text)
    text = text.lower()
    text = text.replace('x', '')
    return text
```

```python
import re
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from nltk.corpus import stopwords
from bs4 import BeautifulSoup
```

```python
def apply_regex(text):
    text = BeautifulSoup(text, "lxml").text
    text = text.lower()

    #Folder path
    text = re.sub(r'[a-z]:\\\S+', r' ', text) 
    #non-ASCII
    text = re.sub(r'[^\x00-\x7F]+', r' ', text)
    #ponctuation
    text = re.sub(r',|\.|\(|\)|\:|\-|\_|\&|\?|\*|\>|\<', r' ', text)
    #Number
    text = re.sub(r'\d+', r' ', text)
    #one letter
    text = re.sub(r' [a-z] ', r' ', text)
    #ntlk stopwords
    text = re.sub(r'\b(' + r'|'.join(stopwords.words('english')) + r')\b\s*', r' ', text)
    
    
    return text
```

```python
def apply_tokenize(text):
    tokens = []
    for sent in nltk.sent_tokenize(text):
        for word in nltk.word_tokenize(sent):
            if len(word) <= 2:
                continue
            tokens.append(word.lower())
    return tokens
```

```python
# Lemmatize with POS Tag
from nltk.corpus import wordnet

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

def apply_lemmatize(text):
    from nltk.stem import WordNetLemmatizer 
    lemmatizer = WordNetLemmatizer()
    text = [lemmatizer.lemmatize(t,get_wordnet_pos(t)) for t in text] 
    return text
```

```python
def apply_stopwordsLocal(text, stopw):
    text =  [t for t in text if t not in stopw]
    return text
```

```python
stopwordsLocal = ['inc','msc','net']
```

```python
def cleanText(text):
    text = apply_regex(text)
    text = apply_tokenize(text)
    text = apply_lemmatize(text)
    text = apply_stopwordsLocal(text, stopwordsLocal)
    return text
```

## training the model

We'll instantiate a Doc2Vec model-Distributed Bag of Words (DBOW). In the Word2Vec architecture, the two algorithm names are “continuous bag of words” (cbow) and “skip-gram” (sg); in the Doc2Vec architecture, the corresponding algorithms are “distributed bag of words” (dbow) and “distributed memory” (dm).

4  DBOW
DBOW is the Doc2Vec model analogous to Skip-gram model in Word2Vec. The paragraph vectors are obtained by training a neural network on the task of predicting a probability distribution of words in a paragraph given a randomly-sampled word from the paragraph.

Training a Doc2Vec model is rather straight forward in Gensim, we initialize the model and train for 30 epochs:

dm =0 means ‘distributed bag of words’ (DBOW), set min_count=2 means ignoring all words with total frequency lower than this, size=100 is dimensionality of the generated feature vectors, alpha=0.025 is the initial alpha rate, learning rate will linearly drop to min_alpha as training progresses. And then we build a vocabulary.

```python
def vec_for_learning(model, tagged_docs):
    sents = tagged_docs.values
    targets, regressors = zip(*[(doc.tags[0], model.infer_vector(doc.words, steps=20)) for doc in sents])
    return targets, regressors
```

```python
def pred_rca(word_extract):
    # load model 
    model_loaded = load_pickle("tmp.model")
    #model_loaded.wv.vocab
    df_x = pd.DataFrame({'Number':['IST0'],'problem_id':[0],'text':[word_extract],'root_cause':['x']})
    from vectorization.data_ingestion.deviation.documentset_factory import make_deviation_set
    sim.deviations = make_deviation_set(df_x, sim.config)
    bow = spacy_bagger.bag(sim.deviations)
    interesting_words = [set([word.lexeme.lower() for word in bow.words])
                             for bow in bow.bows]
    #interesting_words = interesting_words[0]
    #vectorize
    interesting_words = interesting_words[0]
    size = len(interesting_words)
    start = np.zeros(model_loaded.vector_size)
    failures = 0
    for word in interesting_words :
        try :
            start += model_loaded.wv.get_vector(word)
        except :
            failures += 1
    vector = start/size
    vector = pd.DataFrame(vector).T
    #vector
    # load the xgboost model from disk
    filename = 'xgboost_saved.sav'
    xgboost_saved = pickle.load(open(filename, 'rb'))
    result = xgboost_saved.predict(vector)
    result2 = le.inverse_transform (result)
    
    res_proba = np.max(xgboost_saved.predict_proba(vector), axis=1)

    
    return result2[0], res_proba[0]
```
